```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
```

```python
df['Size']=df['Size'].astype(float)
```

```python
df.info()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
item_to_clean=['+',',','$']
cols_to_clean=['Price','Installs']
for items in item_to_clean:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(items,'')
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
df['Price']=df['Price'].astype(float)
df['Installs']=df['Installs'].astype(int)
```

```python
df.info()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
```

```python
df.head(2)
```

```python
df['day']=df['Last Updated'].dt.day
df['month']=df['Last Updated'].dt.month
df['year']=df['Last Updated'].dt.year
```

```python
df.head(2)
```

```python
df.drop('Last Updated',inplace=True,axis=1)
```

```python
df.drop_duplicates('App',inplace=True)
```

```python
numeric_features=[]
categorical_features=[]
for cols in df.columns:
    if df[cols].dtype!='O':
        numeric_features.append(cols)
    else:
        categorical_features.append(cols)
```

```python
numeric_features
```

```python
categorical_features
```

```python
## if you want to plot the numeric feature then this code is necessary
plt.figure(figsize=(15,15))
for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True,color='r',fill=True)
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
