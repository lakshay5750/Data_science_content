```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,random_state=42,n_classes=2)
```

```python
X
```

```python
y
```

```python
from sklearn .model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
penalty=['elasticnet','l2','l1']
c_values=[100,10,1,0.1,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
param=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
param
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=param,scoring='accuracy',n_jobs=-1,cv=cv)
grid.fit(X_train,y_train)
```

```python
grid
```

```python
grid.best_params_
```

```python
grid.score
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)
```

```python
print(score)
print(cm)
print(report)
```
