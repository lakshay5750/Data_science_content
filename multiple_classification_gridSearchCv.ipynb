```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,n_classes=3,n_informative=3,random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
penalty=['l1', 'l2', 'elasticnet']
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
c_values=[100,10,1,0.1,0.01]
multi_class=['auto', 'ovr', 'multinomial']
param=dict(penalty=penalty,solver=solver,C=c_values,multi_class=multi_class)


```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=param,cv=cv,scoring='accuracy',n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)
print(score)
print(cm)
print(report) 
```

```python
regression=LogisticRegression(multi_class='ovr')
```

```python
regression.fit(X_train,y_train)
y_preds=regression.predict(X_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_preds)
cm=confusion_matrix(y_test,y_preds)
report=classification_report(y_test,y_preds)
print(report) 
print(score)
print(cm)
```

```python
##grid seach randomised cv
from sklearn.model_selection import RandomizedSearchCV
random=RandomizedSearchCV(estimator=model,n_jobs=-1,cv=cv,scoring='accuracy',param_distributions=param)
```

```python
random.fit(X_train,y_train)
y_predss=random.predict(X_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_predss)
cm=confusion_matrix(y_test,y_predss)
report=classification_report(y_test,y_predss)
print(report) 
print(score)
print(cm)
```
